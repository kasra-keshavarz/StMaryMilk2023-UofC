```python
import pandas as pd
import pyarrow.feather as feather
import geopandas as gpd
```

```python
# Replace 'your_file_path.feather' with the path to your .feather file
df = feather.read_feather('./OneDrive_1_2-8-2024/SMM_PRMS_HRUs_SCA.feather')
```

```python
# Set the 'datetime' column as the index
df.set_index('datetime', inplace=True)
```

```python
# Remove 'nhmHRU' prefix and convert the remaining part to an integer for each column name
df.columns = [int(col.replace('nhmHRU', '')) for col in df.columns]
```

```python
df
```

```python
gdf = gpd.read_file('./PRMS_validated_shapefile/PRMS_corrected_shapefile.shp')
```

```python
gdf
```

```python
# Create a mapping dictionary from 'hru_nhm' to 'seg_nhm'
mapping_dict = dict(zip(gdf['nhm_id'], gdf['model_hru_']))
```

```python
# Rename columns in the DataFrame using the mapping dictionary
df.rename(columns=mapping_dict, inplace=True)
```

```python
# Convert row index to datetime
df.index = pd.to_datetime(df.index)

# Trim anything later than 2015-12-31
df = df[df.index <= '2015-12-31']
```
